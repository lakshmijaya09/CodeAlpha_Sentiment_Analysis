```python
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt

# Customer Reviews Dataset
reviews = [
    "Product is amazing and works perfectly",
    "Very bad quality and poor performance",
    "Average product, nothing special",
    "I love this item and recommend it",
    "Worst experience ever",
    "Excellent service and fast delivery",
    "Not good, disappointed",
    "Fantastic experience",
    "The product is okay",
    "Great value for money",
    "Terrible packaging",
    "I am happy with the purchase",
    "The quality is poor",
    "Highly recommended product",
    "Delivery was delayed"
]

# Create DataFrame
df = pd.DataFrame(reviews, columns=["Review"])

# Sentiment Analysis Function
def analyze_sentiment(text):
    polarity = TextBlob(text).sentiment.polarity

    if polarity > 0:
        return "Positive"
    elif polarity < 0:
        return "Negative"
    else:
        return "Neutral"

# Emotion Detection Function
def detect_emotion(text):
    text = text.lower()

    if any(word in text for word in ["love", "excellent", "great", "happy", "fantastic"]):
        return "Happy"
    elif any(word in text for word in ["bad", "poor", "worst", "terrible", "disappointed"]):
        return "Angry/Sad"
    else:
        return "Neutral"

# Apply Analysis
df["Sentiment"] = df["Review"].apply(analyze_sentiment)
df["Emotion"] = df["Review"].apply(detect_emotion)

# Display Results
print("========== SENTIMENT ANALYSIS RESULTS ==========")
print(df)

# Sentiment Summary
sentiment_count = df["Sentiment"].value_counts()

print("\n========== SENTIMENT SUMMARY ==========")
print(sentiment_count)

# Emotion Summary
emotion_count = df["Emotion"].value_counts()

print("\n========== EMOTION SUMMARY ==========")
print(emotion_count)

# Sentiment Visualization
plt.figure(figsize=(8,5))
sentiment_count.plot(kind="bar")
plt.title("Sentiment Analysis of Customer Reviews")
plt.xlabel("Sentiment")
plt.ylabel("Number of Reviews")
plt.show()

# Emotion Visualization
plt.figure(figsize=(8,5))
emotion_count.plot(kind="bar")
plt.title("Emotion Detection from Reviews")
plt.xlabel("Emotion")
plt.ylabel("Number of Reviews")
plt.show()

# Conclusion
print("\n========== CONCLUSION ==========")
print("Sentiment analysis helps understand customer opinions.")
print("Positive reviews indicate customer satisfaction.")
print("Negative reviews highlight areas for improvement.")
print("The results can support marketing and product development decisions.")
```
